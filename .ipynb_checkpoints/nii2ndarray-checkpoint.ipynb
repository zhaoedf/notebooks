```python
import numpy as np
import cv2
import SimpleITK as sitk
import os
import matplotlib.pyplot as plt
```

```python
file_names = [name[-8:] for name in os.listdir('../pancreas/images/')]
file_names[:5]
```

```python
len(file_names)
```

```python
index = 0
count = 0
zero_index = 100

pancreas_dataset_path = '../pancreas/pancreas_dataset'
if os.path.exists(pancreas_dataset_path):
    import shutil
    shutil.rmtree(pancreas_dataset_path, ignore_errors=True)
    print("deleting... Done!")
    
os.mkdir(pancreas_dataset_path)
os.mkdir(os.path.join(pancreas_dataset_path,'imgs'))
os.mkdir(os.path.join(pancreas_dataset_path,'masks'))

#file_names = file_names[:5]
for i,name in enumerate(file_names):
    cts = sitk.ReadImage(os.path.join('../pancreas/images/','PANCREAS_'+name))
    labels = sitk.ReadImage(os.path.join('../pancreas/labels/','label'+name))
    
    cts_array = sitk.GetArrayFromImage(cts)
    labels_array = sitk.GetArrayFromImage(labels)
    
    start = -1
    end = -1
    for j in range(cts_array.shape[0]):
        img = cts_array[j]
        mask = labels_array[j]
        if start == -1 and mask.sum() != 0:
            start=j
            continue
        if start != -1 and mask.sum() != 0:
            end = j
        
    print('****************************patient: %d, start: %d, end:%d'%(i,start,end))
#     start = start - 20
#     end = end + 30
    '''
        没必要将那些全是背景的图片加入数据集
        因为单单是含有pancreas的图片中，背景就足够多了！
    '''
    for j in range(cts_array.shape[0]):
        #print(cts_array.shape)
        img = cts_array[j]
        mask = labels_array[j]
        if j < start or j > end: 
            #print(j) 根据需要选择是否打印idx
            continue
            
        
        
        mask = mask*255
        
        print("patient: %d, seq: %d, img_index: %d" %(i,j,index))
        cv2.imwrite(os.path.join(pancreas_dataset_path,'imgs',str(index)+'.jpg'),img)
        cv2.imwrite(os.path.join(pancreas_dataset_path,'masks',str(index)+'_mask.jpg'),mask)
        #print(img.shape)
        #print(mask.sum(),mask.min(),mask.max(),mask.dtype)
        #plt.imshow(img,cmap='gray')
        #plt.imshow(mask,cmap='gray')
        #plt.colorbar()
        
        index += 1

    print('--------------------------------------------------------')
```
